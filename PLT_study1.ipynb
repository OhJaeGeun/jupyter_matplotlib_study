## 2. 밑 그림 그리기

```python
data = np.arange(1, 100)
plt.plot(data)
plt.show() # <-- 그래프를 보여주는 코드.
```

```python
data = np.arange(1, 51)
plt.plot(data)

data2 = np.arange(51, 101)
plt.plot(data2)

plt.show()
```

```python
# figure()는 새로운 그래프 canvas를 생성

data = np.arange(100, 201)
plt.plot(data)

data2 = np.arange(200, 301)
plt.figure()
plt.plot(data2)

plt.show()
```

```python
# 여러개의 plot을 그리는 방법
# subplot(row, column, index)

data = np.arange(100, 201)
plt.subplot(2, 1, 1)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2, 1, 2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(211)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
# 1행 3열짜리 그래프 3개를 작성
# 1행 1열 : 100 ~ 200, 1행 2열 : 200 ~ 300, 1행 3열 : 300 ~ 400

data = np.arange(100, 201)
plt.subplot(131)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(132)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(133)
plt.plot(data3)

plt.show()
```

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 6])

# 타이틀
plt.title(('타이틀 테스트'), fontsize = 30)

# x, y축 label 설정
plt.xlabel('x축', fontsize = 20)
plt.ylabel('y축', fontsize = 20, rotation = 0)

plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title('범례 설정', fontsize = 30)

plt.xlabel('x축', fontsize = 20)
plt.ylabel('y축', fontsize = 20, rotation = 0)

plt.xticks(rotation = 90)
plt.yticks(rotation = 30)

plt.legend(['10*2', '10**2', 'log'], fontsize = 15, loc = 'upper right')

# x와 y의 한계점(limit) 설정
plt.xlim(0, 10)
plt.ylim(0.5, 20)

plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2, marker = 'o', markersize = 5, linestyle = '', color = 'b', alpha = 0.3)
plt.plot(np.arange(10), np.arange(10)**2, marker = 'v', markersize = 10, linestyle = '--', color = 'c', alpha = 0.5)
plt.plot(np.arange(10), np.log(np.arange(10)), marker = '*', markersize = 15, linestyle = '-.', color = 'y', alpha = 0.7)

plt.title('마커 설정', fontsize = 30)

plt.xlabel('x축', fontsize = 20)
plt.ylabel('y축', fontsize = 20, rotation = 0)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

# 그리드 설정
plt.grid()

plt.show()
```

```python
x = np.random.rand(50) # 0 ~ 1 소수 50개
y = np.random.rand(50)

colors = np.arange(50)

area = x * y * 10000

print(x)
print(y)
print(colors)
print(area)
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
# cmap에 컬러를 설정하면 모든 그래프에 색상을 같게 설정해줌
plt.figure(figsize = (12, 6))

plt.subplot(131)
plt.scatter(x, y, s = area, c = 'blue', alpha = 0.1)
plt.title('alpha = 0.1')

plt.subplot(132)
plt.scatter(x, y, s = area, cmap = 'blue', alpha = 0.5)
plt.title('alpha = 0.5')

plt.subplot(133)
plt.scatter(x, y, s = area, cmap = 'blue', alpha = 1.0)
plt.title('alpha = 1.0')

plt.show()
```

```python
# Line Plot

x = np.arange(0, 10, 0.1)
y = 5 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize = 15)
plt.ylabel('y value', fontsize = 15)
plt.title('sine graph', fontsize = 30)

plt.grid()

plt.show()
```

```python
# pie chart
labels = ['애플', '마이크로소프트', 'IBM', '나이키', '에비디아', 'AT&T']
sizes = [20.5,     15.8,            10.5,   9,        7.6,        36.6]

explode = (0.3, 0, 0, 0.9, 0, 0)

# 그냥 파이차트는 이렇게 쓰는 것인듯??
# plt.pie로 생기는 요소를 다음처럼 리턴하여 값을 저장
patches, texts, autotexts = plt.pie(
    sizes,
    explode = explode, # 강조느낌(원에서 떨어져 나감)
    labels = labels, # 숫자에 맞는 회사 표시
    autopct = '%1.1f%%', # 숫자 뒤에 퍼센트 표시
    shadow = True, # 그림자
    startangle = 90 # ???
)

plt.title('성공의 지름길', fontsize = 30)

# label 텍스트에 대한 스타일 적용
for t in texts:
    t.set_fontsize(13)
    t.set_color('gray')
    
# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
    t.set_color('white')
    t.set_fontsize(17)
    
plt.show()
```
